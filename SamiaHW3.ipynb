```
#SamiaHW3
#AnsToTheQuestion1
from numpy import *
import matplotlib.pyplot as plt
from scipy.integrate import odeint # to solve differential equation for population


```

```
#Defining parameters
betaR=0.01
deltaR=0.005
betaF=0.005
deltaF=0.01
epsilonR=0.0001
epsilonF=epsilonR
R0=100
F0=100
P0=array([R0,F0])


```

```
def pop(P,t):
  return array([betaR*P[0]-deltaR*P[0]-epsilonF*P[0]*P[1], betaF*P[1]-deltaF*P[1]+epsilonR*P[0]*P[1]]) #function using differential equation of population
```

```
t = linspace(0, 10000, 1000)  # time 
P= odeint(pop,P0,t)           # Population
R,F = P.T  #Transpose

  
```

```
#Plotting the population growth 
  p= plt.figure(figsize=(10,6)) 
  plt.plot(t, R, 'b-', label='Rabbits') 
  plt.plot(t, F, 'r-', label='Foxes') 
  plt.legend(loc='upper right')
  plt.grid() 
  plt.xlabel('time (days)') 
  plt.ylabel('population') 
  plt.title('Evolution of population')
```

```
#AnsToTheQuestion3(a)(b)
import numpy as np
import math
import matplotlib.pyplot as plt
U_235_Pb_207 = np.array([4321,4303,4153,3155,3672,2808,4355,4307,4078,4287,3651,3317,4392,3985]) #207Pb/235U (Myr)
U_238_Pb_206 = np.array([4233,4201,3780,1713,2595,1177,4356,4235,3674,4126,2604,1961,4364,3355]) #206Pb/238U (Myr)
```

```
U_235_Pb_207 = U_235_Pb_207*(math.pow(10,6)) 
U_238_Pb_206 = U_238_Pb_206*(math.pow(10,6)) 
```

```
ts = np.linspace(0,6000,100) 
t = ts *(math.pow(10,6)) 
lambda_235 = 9.85 * (10**(-10)) #half-life 
lambda_238 = 1.54 * (10**(-10)) 
```

```
Concordia_235=(np.exp(lambda_235*t)) - 1
Concordia_238=(np.exp(lambda_238*t)) - 1

Discordia_235 =(np.exp(lambda_235*U_235_Pb_207)) - 1 
Discordia_238 =(np.exp(lambda_238*U_238_Pb_206)) - 1 
```

```
z = np.polyfit(Discordia_235,Discordia_238, 1) 
p = np.poly1d(z)

xp = np.linspace(-1, 120, 1000) 
 
fig1 = plt.figure(figsize=(10,8)) 
plt.plot(Concordia_235 , Concordia_238, 'g-',label = 'Concordia') 
plt.plot(Discordia_235 , Discordia_238, 'b.', label = 'Discordia Data')
plt.plot(xp, p(xp), 'r-',label='Discordia')
plt.grid() 
plt.legend(loc='lower right') 
plt.ylabel('206Pb/238U') 
plt.xlabel('207Pb/235U') 
plt.title('Concordia and Discordia')  
```

```
#AnsToTheQuestion5(a)(b)(c)
import numpy as np
import matplotlib.pyplot as plt

Sr_Sr = np.array([0.7117,0.7095,0.7092,0.7083,0.7083,0.7082]) #87Sr/86Sr ratio
Rb_Sr = np.array([3.65,1.80,1.48,0.82,0.66,0.74]) #87Rb/86Sr ratio 
z = np.polyfit(Rb_Sr,Sr_Sr, 1)
p = np.poly1d(z) 

fig = plt.figure(figsize=(10,8))
xp = np.linspace(0, 6, 1000) 
plt.plot(Rb_Sr , Sr_Sr, 'bo',label = 'Rb-Sr isochron data point') 
plt.plot(Rb_Sr, Sr_Sr, '.', xp, p(xp), 'r-') 
plt.legend(loc='lower right')
plt.ylabel('87Sr/86Sr') 
plt.xlabel('87Rb/86Sr') 
plt.title('Rb-Sr isochron')
plt.grid() 
print(p)  
```
